### df.isin()

```python
data = [['Dan', '   Marketing'],
        ['Paul', 'Programming'],
        ['Helen', 'Marketing'],
        ['Donald', 'Accounting'],
        ['Diana', 'Networking']]
my_df = pd.DataFrame(data=data, columns=['Name', 'Skills'])
```

```python
my_df
```

```python
skills = my_df['Skills'].str.strip().isin(['Programming', 'Marketing'])
```

```python
type(skills)
```

```python
skills
```

```python
my_df[skills]
```

### df.largest() and df.nsmallest()

```python
df.nlargest(5, 'GDP ($ per capita)')
```

```python
df.nlargest(n=2, columns='Birthrate')
```

```python
df.nsmallest(5, 'Area (sq. mi.)').sort_values('Population', ascending=False)
```
